```python
import albumentations as albu
from skimage import io
import cv2
import os
import numpy as np
import PIL
```

```python
images = os.listdir('D:/Practice_Code/Gauss_noise/Test_noise/Data_org')
pathin = 'D:/Practice_Code/Gauss_noise/Test_noise/Data_org/'
pathout = 'D:/Practice_Code/Gauss_noise/Test_noise'
noise1 = 0
noise2 = 250

for img in images:
    for i in [noise1,noise2]:
         if(i==noise1):
#             image = io.imread('D:/Practice_Code/Gauss_noise/Test_noise/Data_org/' + img)
            image = io.imread(pathin + img)
            transform =albu.GaussNoise(var_limit=(noise1,noise1),mean=0,p=0.5)
            image = transform(image=image)['image']
            PIL.Image.fromarray(image).save(pathout + '/Noise1/{}'.format(img))
         elif(i==noise2):
           # image = io.imread('D:/Practice_Code/Gauss_noise/Test_noise/Data_org/' + img)
            image = io.imread(pathin + img)
            transform =albu.GaussNoise(var_limit=(noise2,noise2),mean=0,p=0.5)
            image = transform(image=image)['image']
            PIL.Image.fromarray(image).save(pathout+'/Noise2/{}'.format(img))
```

```python
# Code thống kê so sánh ảnh trước và sau khi gây nhiễu
import json
import os
from collections import namedtuple
from json import JSONEncoder
```

```python
images = os.listdir('D:/Image_Test/')
pathout='D:/Label/'

with open(pathout + 'Data_org.json', 'r') as f:
    data = json.load(f)
with open(pathout + 'Data_1.json', 'r') as n1:
    data1 = json.load(n1)
with open(pathout + 'Data_2.json', 'r') as n2:
    data2 = json.load(n2)

temp1_n1 = 0
temp2_n1 = 0
temp3_n1 = 0
temp1_n2 = 0
temp2_n2 = 0
temp3_n2 = 0


for img in images:
    temp1_n1 +=1
    if data[img] == data1[img]:
        temp2_n1 +=1
for img in images:
    if data[img] == "" and data1[img] == "":
        temp3_n1 +=1

for img in images:
    temp1_n2 +=1
    if data[img] == data2[img]:
        temp2_n2 +=1
for img in images:
    if data[img] == "" and data2[img] == "":
        temp3_n2 +=1

with open(pathout+"/Datafinal.txt", mode='a',  encoding="utf-8") as fff:
    fff.writelines(f'\n=========================================================')
    fff.writelines(f'\n+++NOISE1+++++++++++++++++++++')
    fff.writelines(f'\nTong so anh xư ly: {temp1_n1}')
    fff.writelines(f'\nSo anh nhan dien giong khi chua pha nhieu: {temp2_n1}')
    fff.writelines(f'\nKy tu trong: {temp3_n1}')
    fff.writelines(f'\nTy le % so anh nhan dien giong khi chua pha nhieu: {float(temp2_n1/(temp1_n1)*100)}%')
    fff.writelines(f'\nTy le % so anh nhan dien giong khi chua pha nhieu(tru ky tu trong): {float((temp2_n1-temp3_n1)/(temp1_n1)*100)}%')
    fff.writelines(f'\n+++NOISE2+++++++++++++++++++++')
    fff.writelines(f'\nTong so anh xư ly: {temp1_n2}')
    fff.writelines(f'\nSo anh nhan dien giong khi chua pha nhieu: {temp2_n2}')
    fff.writelines(f'\nKy tu trong: {temp3_n2}')
    fff.writelines(f'\nTy le % so anh nhan dien giong khi chua pha nhieu: {float(temp2_n2/(temp1_n2)*100)}%')
    fff.writelines(f'\nTy le % so anh nhan dien giong khi chua pha nhieu(tru ky tu trong): {float((temp2_n2-temp3_n2)/(temp1_n2)*100)}%')

print(f'+++Noise1++++++++++++')
print(f'Tong so anh xư ly: {temp1_n1}')
print(f'So anh nhan dien giong khi chua pha nhieu: {temp2_n1}')
print(f'Ky tu trong truoc va sau nhan dien: {temp3_n1}')
print(f'Ty le % so anh nhan dien giong khi chua pha nhieu: {float(temp2_n1/(temp1_n1)*100)}%')
print(f'+++Noise2++++++++++++')
print(f'Tong so anh xư ly: {temp1_n2}')
print(f'So anh nhan dien giong khi chua pha nhieu: {temp2_n2}')
print(f'Ky tu trong truoc va sau nhan dien: {temp3_n2}')
print(f'Ty le % so anh nhan dien giong khi chua pha nhieu: {float(temp2_n2/(temp1_n2)*100)}%')
```
